```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
datos = pd.read_excel(
    "../data/reservoir/LaFE_Elevacion-Area-Volumen.xlsx",
    skiprows=8)
datos.columns = ["Nivel", "Area", "VolUtil", "VolTotal"]
curva = datos[["Nivel", "VolUtil"]].dropna()
curva = curva.sort_values(by="Nivel")

f_level_to_volume = CubicSpline(curva.Nivel, curva.VolUtil)
f_volume_to_level = CubicSpline(curva.VolUtil, curva.Nivel)
```

```python
plt.plot(curva.Nivel, curva.VolUtil)
```
